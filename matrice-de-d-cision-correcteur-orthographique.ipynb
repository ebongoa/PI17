# **Temps d'éxécution, RAM, CPU**

```python
X = ['autocorrect', 'pyspellchecker', 'textblob', 'enchant']
times = []
cpu = []
memory = []
```

```python
import pytesseract
from PIL import Image
```

```python
import re
```

```python
texte = pytesseract.image_to_string(Image.open('../input/texte-numrique/largepreview.png'))
str = re.findall("[abcdefghijklmnopqrstuvwxzyãáàâçéêíõóôúü,A-Z]+",texte)
text = (" ".join(str))
    
print(text)
```

```python
from __future__ import print_function
import time
import psutil
import os
```

## **Autocorrect**

```python
from autocorrect import Speller
import autocorrect
```

```python
spell = Speller('fr')
correct = ""
start = time.time()

for line in text.splitlines():
    add = spell.autocorrect_sentence(line)
    correct = correct + add + ' '    
print(correct)

end = time.time()
times.append(end-start)
cpu.append(psutil.cpu_percent())
pid = os.getpid()
py = psutil.Process(pid)
memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
```

## **PySpellChecker**

```python
from spellchecker import SpellChecker
```

```python
spell = SpellChecker(language='fr', distance=2)
correct = ""
start = time.time()

for word in text.split():
    add = spell.correction(word)
    correct = correct + add + ' '    
print(correct)

end = time.time()
times.append(end-start)
cpu.append(psutil.cpu_percent())
pid = os.getpid()
py = psutil.Process(pid)
memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
```

## **TextBlob**

```python
from textblob import TextBlob
```

```python
spell = TextBlob(text)
start = time.time()

correct = spell.correct().string
print(correct)

end = time.time()
times.append(end-start)
cpu.append(psutil.cpu_percent())
pid = os.getpid()
py = psutil.Process(pid)
memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
```

## **Enchant**

```python
import enchant
```

```python
spell = enchant.Dict('fr')
correct = ''
start = time.time()

for word in text.split():
    add = spell.suggest(word)
    if len(add)!=0 :
        correct = correct + add[0] + ' '
    else :
        correct = correct + word + ' '
print(correct)

end = time.time()
times.append(end-start)
cpu.append(psutil.cpu_percent())
pid = os.getpid()
py = psutil.Process(pid)
memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
```

## **Visualisation**

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize = (15, 8))

model_comparison = pd.DataFrame({'model': X,'time': times, 'cpu': cpu, 'memory': memory})
model_comparison.sort_values('time', ascending = False).plot(x = 'model', y = ['time', 'cpu', 'memory'], kind = 'barh', color = 'rgb', edgecolor = 'black')

plt.ylabel('')
plt.yticks(size = 10)
plt.xlabel('')
plt.xticks(size = 10)
plt.title('Comparaison des solutions', size=20)
print(X)
print(times)
print(cpu)
print(memory)
```
