## **Texte Numérique**

```python
import pytesseract
import pyocr
import easyocr
```

```python
import time
from PIL import Image
```

```python
tools = pyocr.get_available_tools()
# The tools are returned in the recommended order of usage
tool = tools[0]
```

```python
X = ['Tesseract', 'PyOCR', 'EasyOCR']
times = []
memory = []
cpu = []
```

```python
from pytesseract import Output
```

```python
from __future__ import print_function
import psutil
import os
```

```python
start = time.time()
text = pytesseract.image_to_data(Image.open('../input/texte-numrique/largepreview.png'), output_type=Output.DICT)
print(text['conf'])
print(text['text'])
end = time.time()
times.append(end-start)
cpu.append(psutil.cpu_percent())
pid = os.getpid()
py = psutil.Process(pid)
memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
```

```python
start = time.time()
textbuilder = pyocr.builders.TextBuilder()
text = tool.image_to_string(Image.open('../input/texte-numrique/largepreview.png'),builder=textbuilder)
print(text)
end = time.time()
times.append(end-start)
cpu.append(psutil.cpu_percent())
pid = os.getpid()
py = psutil.Process(pid)
memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
```

```python
start = time.time()
reader = easyocr.Reader(['ch_sim','en']) # this needs to run only once to load the model into memory
text = reader.readtext('../input/texte-numrique/largepreview.png')
print(text)
end = time.time()
times.append(end-start)
cpu.append(psutil.cpu_percent())
pid = os.getpid()
py = psutil.Process(pid)
memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize = (15, 8))

model_comparison = pd.DataFrame({'model': X,'time': times, 'cpu': cpu, 'memory': memory})
model_comparison.sort_values('time', ascending = False).plot(x = 'model', y = ['time', 'cpu', 'memory'], kind = 'barh', color = 'rgb', edgecolor = 'black')

plt.ylabel('')
plt.yticks(size = 10)
plt.xlabel('')
plt.xticks(size = 10)
plt.title('Comparaison des solutions', size=20)   
```

```python
start = time.time()
data = pytesseract.image_to_data(Image.open('../input/imagetexte/texte-non-defini-avec-des-mots-anglais-lettre-manuscrite-ecriture-manuscrite-calligraphie-manuscrit-script-police-arriere-plan-de-texture-abstraite-2fpd614.jpg'))
# print(data)
end = time.time()
y.append(end-start)
```
