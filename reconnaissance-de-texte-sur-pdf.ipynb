## D√©tection de la langue dans une image

```python
import pytesseract
import fasttext
import pycountry
from collections import Counter

PRETRAINED_MODEL_PATH = '/tmp/lid.176.bin'
model = fasttext.load_model(PRETRAINED_MODEL_PATH)
```

```python
def code_coef_lang(code, images) :
    # Returns from a language code and the images,
    # the language code of the text present in the images with the certainty coefficient.
    text = ''
    for image in images :
        text = text + pytesseract.image_to_string(image, lang=code) + ' '
    # print(text.replace('\n', ' '))
    label = model.predict(text.replace('\n', ' '))[0][0]
    code1 = label.replace("__label__", '')
    lang = pycountry.languages.get(alpha_2=code1)
    code1 = lang.alpha_3
        
    if 'zho' in code1:
        code1 = 'chi_sim'
        
    coef = model.predict(text.replace('\n', ' '))[1][0]
        
    if code1 != code :
        return code_coef_lang(code1, images)
    else :
        return code, coef
```

```python
def detect_images_lang(images) :
    # Detects the most likely language of the text in the images.
    # Our language base is the one on the MyBrian platform.
    # This database will need to be updated if it changes.
    coefs = []
    codes = []
    langs = ['eng', 'fra', 'spa','chi_sim', 'deu', 'ita', 'por', 'rus', 'jpn', 'nld', 'ell', 'pol', 'swe', 'ukr']
    
    for lang in langs :
        code, coef = code_coef_lang(lang, images)
        # print(code)
        codes.append(code)
        # print(coef)
        coefs.append(coef)
        if coef>0.90 :
            break
                    
    if Counter(codes).most_common(1)[0][1]>1 :
        code = Counter(codes).most_common(1)[0][0]
    else : 
        coef = max(coefs)
        index = coefs.index(coef)
        code = codes[index]

    # print(code, coef)
    return code
```

## Transformation des images obtenues en un texte

```python
from PIL import Image
import pytesseract
```

```python
def images_to_text(images):
    # Transforms the list of images into a text.
    lang = detect_images_lang(images)
    custom_config = r'--psm 6'
    text = ''
    for image in images : 
        text = text + pytesseract.image_to_string(image, lang=lang, config=custom_config) + '\n'
    return text  
```

## Transformation du pdf en texte

```python
import fitz
import pdfplumber 
from PIL import Image
import io
import tempfile
from pdf2image import convert_from_path
```

```python
def pdf_to_text(pdf_path) :
    # Convert the pdf file into text 
    # whether it is a scan pdf or a simple pdf.
    text = ''
    with pdfplumber.open(pdf_path) as pdf:
        # Extract text from pdf.
        for page in pdf.pages :
            if page.extract_text() != '':
                text = text + page.extract_text() + '\n'
    # print(text)
    if text == '':
        # If the pdf has no text, it is a scan.
        with tempfile.TemporaryDirectory() as path:
            images_from_path = convert_from_path(pdf_path, output_folder=path)
            text = images_to_text(images_from_path)
        # print(text)
        return text
    else : 
        ## The pdf can be complex i.e. contain text and images.
        # with fitz.open(pdf_path) as pdf:
        #    images = []
        #    for page in pdf.pages():
        #        list_image = page.getImageList()
        #        if len(list_image)!=0 :
        #            # print(list_image)
        #            for image_number, img in enumerate(list_image):
        #                xref = img[0]
        #                image_base = pdf.extractImage(xref)
        #                # print(image_base)
        #                bytes_image = image_base["image"]
        #                image = Image.open(io.BytesIO(bytes_image))
        #                images.append(image)
        #    if len(images) != 0:
        #        text = text + '\n' + images_to_text(images)
        # print(text)
        return text
```
