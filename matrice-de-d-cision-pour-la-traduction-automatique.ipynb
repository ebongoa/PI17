## **Traduction de la langue**

```python
from textblob import TextBlob
import goslate
from googletrans import Translator
```

```python
from __future__ import print_function
import time
import psutil
import os
```

```python
texts = [fra, mex, deu, esp, ita, nl, pl, sv, zh, jap]
langues = ["fra", "mex", "deu", "esp", "ita", "nl", "pl", "sv", "zh", "jap"]
# texts
```

#### **TextBlob**

```python
trads_textblob = []
times = []
memory = []
cpu = []

for text in texts :
    start = time.time()
    trad_textblob = TextBlob(text)
    trads_textblob.append(trad_textblob.translate(to='en'))
    end = time.time()
    times.append(end-start)
    cpu.append(psutil.cpu_percent())
    pid = os.getpid()
    py = psutil.Process(pid)
    memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
    
print(texts)
print(trads_textblob)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize = (15, 8))

model_comparison = pd.DataFrame({'langue': langues,'time': times, 'cpu': cpu, 'memory': memory})
model_comparison.sort_values('time', ascending = False).plot(x = 'langue', y = ['time', 'cpu', 'memory'], kind = 'barh', color = 'rgb', edgecolor = 'black')

plt.ylabel('')
plt.yticks(size = 10)
plt.xlabel('')
plt.xticks(size = 10)
plt.title('Comparaison de TextBlob avec les langues', size=20)                            
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize = (15, 8))

model_comparison = pd.DataFrame({'langue': langues,'time': times})
model_comparison.sort_values('time', ascending = False).plot(x = 'langue', y ='time', kind = 'barh', color = 'blue', edgecolor = 'black')

plt.ylabel('')
plt.yticks(size = 10)
plt.xlabel('Time')
plt.xticks(size = 10)
plt.title('Comparaison de TextBlob avec les langues', size=20)   
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize = (15, 8))

model_comparison = pd.DataFrame({'langue': langues,'cpu': cpu})
model_comparison.sort_values('cpu', ascending = False).plot(x = 'langue', y='cpu', kind = 'barh', color = 'orange', edgecolor = 'black')

plt.ylabel('')
plt.yticks(size = 10)
plt.xlabel('CPU')
plt.xticks(size = 10)
plt.title('Comparaison de TextBlob avec les langues', size=20)   
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize = (15, 8))

model_comparison = pd.DataFrame({'langue': langues, 'memory': memory})
model_comparison.sort_values('memory', ascending = False).plot(x = 'langue', y ='memory', kind = 'barh', color = 'green', edgecolor = 'black')

plt.ylabel('')
plt.yticks(size = 10)
plt.xlabel('Memory')
plt.xticks(size = 10)
plt.title('Comparaison de TextBlob avec les langues', size=20)   
```

#### **Goslate**

```python
trads_goslate = []
times = []
memory = []
cpu = []

for text in texts :
    start = time.time()
    gs = goslate.Goslate()
    trads_goslate.append(gs.translate(text, 'en'))
    end = time.time()
    times.append(end-start)
    cpu.append(psutil.cpu_percent())
    pid = os.getpid()
    py = psutil.Process(pid)
    memory.append(py.memory_info()[0]/2.**30) #  physical memory usage
    
print(texts)
print(trads_goslate)
```
